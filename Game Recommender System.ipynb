```python
df['tags'] = df['publisherName'].astype(str) + df['name'].astype(str) + df['starRating/averageRating'].astype(str)
df
```

```python
df['tags'] = df['publisherName'].astype(str) + ' ' + df['name'].astype(str) + ' ' + df['starRating/averageRating'].astype(str)
print(df['tags'][0])
```

```python
new_df=df[['name', 'tags']]
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=1000)
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
new_df.rename(columns={'name':'title'}, inplace=True)
```

```python
def recommend(ps5):
    game_index=new_df[new_df['name']==ps5].index[0]
    distances=similarity[game_index]
    game_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in game_list:
        print(new_df.iloc[i[0]].title)
```
